```python
# "Lobo frontal: relacionado com trabalho criativo, raciocÃ­nio, personalidade, tomada de decisÃµes, movimentaÃ§Ã£o dos mÃºsculos esquelÃ©ticos, entre outras funÃ§Ãµes.

# Lobo temporal: atua, principalmente, na comunicaÃ§Ã£o, relacionando-se com a audiÃ§Ã£o e compreensÃ£o da linguagem.

# Lobo parietal: relacionado, entre outras funÃ§Ãµes, com a percepÃ§Ã£o de dor, frio, calor e toques.

# Lobo occipital: possui relaÃ§Ã£o com o processamento das informaÃ§Ãµes visuais."

# Veja mais sobre "CÃ©rebro" em: https://brasilescola.uol.com.br/biologia/cerebro-humano.htm



class Cerebro:
    def __init__(self,Fp1,Fp2,F7,F3,Fz,F4,F8,FC5,FC1,FC2,FC6,T7,C3,Cz,C4,T8,TP9,CP5,CP1,CP2,CP6,TP10,P7,P3,Pz,P4,P8,PO9,O1,Oz,O2,PO10):
        self.Fp1=np.array(Fp1)
        self.Fp2=np.array(Fp2)
        self.F7=np.array(F7)
        self.F3=np.array(F3)
        self.Fz=np.array(Fz)
        self.F4=np.array(F4)
        self.F8=np.array(F8)
        self.FC5=np.array(FC5)
        self.FC1=np.array(FC1)
        self.FC2=np.array(FC2)
        self.FC6=np.array(FC6)
        self.T7=np.array(T7)
        self.C3=np.array(C3)
        self.Cz=np.array(Cz)
        self.C4=np.array(C4)
        self.T8=np.array(T8)
        self.TP9=np.array(TP9)
        self.CP5=np.array(CP5)
        self.CP1=np.array(CP1)
        self.CP2=np.array(CP2)
        self.CP6=np.array(CP6)
        self.TP10=np.array(TP10)
        self.P7=np.array(P7)
        self.P3=np.array(P3)
        self.Pz=np.array(Pz)
        self.P4=np.array(P4)
        self.P8=np.array(P8)
        self.PO9=np.array(PO9)
        self.O1=np.array(O1)
        self.Oz=np.array(Oz)
        self.O2=np.array(O2)
        self.PO10=np.array(PO10)

    
    def obterLobos(self):
                
        self.LoboFrontalE=(self.Fp1+self.F7+self.F3)/3
        self.LoboFrontalD=(self.Fp2+self.F4+self.F8)/3
        
        self.LoboParietalE=(self.CP5+self.CP1+self.P3)/3
        self.LoboParietalD=(self.CP2+self.CP6+self.P4)/3
        
        self.LoboTemporalE=(self.T7+self.TP9)/2
        self.LoboTemporalD=(self.T8+self.TP10)/2
        
        self.LoboOccipitalE=O1
        self.LoboOccipitalD=O2

        
        return self.LoboFrontalE,self.LoboFrontalD, self.LoboParietalE,self.LoboParietalD,self.LoboTemporalE,self.LoboTemporalD,self.LoboOccipitalE,self.LoboOccipitalD
    
```

```python
class FiltraCerebro:
    def __init__(self,filtroDelta,filtroTeta,filtroAlfa,filtroBeta,N,T):
        self.filtroDelta=filtroDelta
        self.filtroTeta=filtroTeta
        self.filtroAlfa=filtroAlfa
        self.filtroBeta=filtroBeta
        self.tempo=np.arange(0, T*N, T)
        
    def filtrar(self,lobo,tipoDeFiltro):
        print(".",end="")
        tout, y, x=signal.lsim(filtroDelta,lobo, self.tempo)
        
        if(tipoDeFiltro=="teta"):
            tout, y, x=signal.lsim(filtroTeta,lobo, self.tempo)
        if(tipoDeFiltro=="alfa"):
            tout, y, x=signal.lsim(filtroAlfa,lobo, self.tempo)
        if(tipoDeFiltro=="beta"):
            tout, y, x=signal.lsim(filtroBeta,lobo, self.tempo)
            
        return y   
        
    def obterOndas(self,lobo):
        return self.filtrar(lobo,"delta"),self.filtrar(lobo,"teta"),self.filtrar(lobo,"alfa"),self.filtrar(lobo,"beta")
```

```python
class Relatorio:
    def __init__(self,delta,teta,alfa,beta):
        self.delta=delta
        self.teta=teta
        self.alfa=alfa
        self.beta=beta
    def calcularEnergia(self):

               
        T=1/500
        t=T*len(self.delta)
        energiaDelta=np.sum(np.abs(self.delta**2))/(T)
        energiaTeta=np.sum(np.abs(self.teta**2))/(T)
        energiaAlfa=np.sum(np.abs(self.alfa**2))/(T)
        energiaBeta=np.sum(np.abs(self.beta**2))/(T)
        
        return energiaDelta, energiaTeta, energiaAlfa, energiaBeta           
    def plotarFFT(self,title):
        D,T,A,B=self.calcularEnergia()
        energiaTotal=D+T+A+B
        
        pesoTotal=200+100+50+30
        
        eD=D/energiaTotal
        eT=T/energiaTotal
        eA=A/energiaTotal
        eB=B/energiaTotal
        freq=np.linspace(0,500,len(self.delta))
        deltaFFT=fft(self.delta)/len(self.delta)
        alfaFFT=fft(self.alfa)/len(self.alfa)
        betaFFT=fft(self.beta)/len(self.beta)
        tetaFFT=fft(self.teta)/len(self.teta)
        plt.figure(figsize=[18,5])
        plt.title("Atividade Cerebra-> " +title)
        plt.xlabel("Hz")
        plt.ylabel("Intensidade")
        plt.plot(freq,np.abs(deltaFFT),label="Energia delta -> "+str(100*eD//1)+"%",alpha=0.4)
        plt.plot(freq,np.abs(alfaFFT),label="Energia alfaa -> "+str(100*eA//1)+"%",alpha=0.4)
        plt.plot(freq,np.abs(betaFFT),label="Energia beta -> "+str(100*eB//1)+"%",alpha=0.4)
        plt.plot(freq,np.abs(tetaFFT),label="Energia teta -> "+str(100*eT//1)+"%",alpha=0.4)
        plt.xlim([0, 35])
        plt.legend()
        
    def plotarFFTNormalizada(self,title):
        D,T,A,B=self.calcularEnergia()
        energiaTotal=D+T+A+B
        
        pesoTotal=200+100+50+30
        
        eD=D/energiaTotal
        eT=T/energiaTotal
        eA=A/energiaTotal
        eB=B/energiaTotal
        freq=np.linspace(0,500,len(self.delta))
        deltaFFT=fft(self.delta)/len(self.delta)
        alfaFFT=fft(self.alfa)/len(self.alfa)
        betaFFT=fft(self.beta)/len(self.beta)
        tetaFFT=fft(self.teta)/len(self.teta)
        plt.figure(figsize=[18,5])
        plt.title("Atividade Cerebra-> " +title)
        plt.xlabel("Hz")
        plt.ylabel("Intensidade")
        plt.plot(freq,np.abs(deltaFFT)/200,label="Energia delta -> "+str(100*eD//1)+"%(Amplificado 1 vez)",alpha=0.4)
        plt.plot(freq,np.abs(alfaFFT)/100,label="Energia alfaa -> "+str(100*eA//1)+"%(Amplificado 2 vezes)",alpha=0.4)
        plt.plot(freq,np.abs(betaFFT)/50,label="Energia beta -> "+str(100*eB//1)+"%(Amplificado 4 vezes)",alpha=0.4)
        plt.plot(freq,np.abs(tetaFFT)/30,label="Energia teta -> "+str(100*eT//1)+"%(Amplificado 6 vezes)",alpha=0.4)
        plt.xlim([0, 35])
        plt.legend()        
        
        
    def plotarFase(self,title):
        D,T,A,B=self.calcularEnergia()
        energiaTotal=D+T+A+B
        
        pesoTotal=200+100+50+30
        
        eD=D/energiaTotal
        eT=T/energiaTotal
        eA=A/energiaTotal
        eB=B/energiaTotal
        freq=np.linspace(0,500,len(self.delta))
        deltaFFT=fft(self.delta)/len(self.delta)
        alfaFFT=fft(self.alfa)/len(self.alfa)
        betaFFT=fft(self.beta)/len(self.beta)
        tetaFFT=fft(self.teta)/len(self.teta)
        plt.figure(figsize=[18,5])
        plt.title("Atividade Cerebra-> " +title)
        plt.xlabel("Hz")
        plt.ylabel("Intensidade")
        plt.plot(freq,np.angle(deltaFFT),label="Energia delta -> "+str(100*eD//1)+"%",alpha=0.4)
        plt.plot(freq,np.angle(alfaFFT),label="Energia alfa -> "+str(100*eA//1)+"%",alpha=0.4)
        plt.plot(freq,np.angle(betaFFT),label="Energia beta -> "+str(100*eB//1)+"%",alpha=0.4)
        plt.plot(freq,np.angle(tetaFFT),label="Energia teta -> "+str(100*eT//1)+"%",alpha=0.4)
        plt.xlim([0, 30])
        plt.legend()
        

relatorioLFE= Relatorio(deltaLFE,tetaLFE,alfaLFE,betaLFE)
relatorioLPE= Relatorio(deltaLPE,tetaLPE,alfaLPE,betaLPE)
relatorioLTE= Relatorio(deltaLTE,tetaLTE,alfaLTE,betaLTE)
relatorioLOE= Relatorio(deltaLOE,tetaLOE,alfaLOE,betaLOE)

relatorioLFD= Relatorio(deltaLFD,tetaLFD,alfaLFD,betaLFD)
relatorioLPD= Relatorio(deltaLPD,tetaLPD,alfaLPD,betaLPD)
relatorioLTD= Relatorio(deltaLTD,tetaLTD,alfaLTD,betaLTD)
relatorioLOD= Relatorio(deltaLOD,tetaLOD,alfaLOD,betaLOD)
```
